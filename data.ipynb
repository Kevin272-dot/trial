```python
import csv
import random

# Define list of regions
regions = ['North', 'South', 'East', 'West', 'Central']

# File name
filename = 'simple_car_sales_data.csv'

# Generate data
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(['Region', 'Units_Sold'])
    
    for _ in range(100):  # 100 rows
        region = random.choice(regions)
        units = random.randint(1, 50)  # Random units sold
        writer.writerow([region, units])

print(f"CSV file '{filename}' with 100 rows and 2 columns generated successfully.")
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
iris= load_iris()
x = iris.data
y = iris.target
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Train the classifier
clf.fit(x_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(x_test) 
# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
print("Accuracy:", accuracy)
print("Recall:", recall)
```
